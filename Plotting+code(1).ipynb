```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the S-parameter data
filename = 'C:\\Users\\hp\\Desktop\\New folder\\2.5 S Para.csv'
data = pd.read_csv(filename)

# Extract the data
frequency = data['Frequency (Hz)']
S11 = data['S11 (Real)'] + 1j * data['S11 (Imag)']
S21 = data['S21 (Real)'] + 1j * data['S21 (Imag)']
S12 = data['S12 (Real)'] + 1j * data['S12 (Imag)']
S22 = data['S22 (Real)'] + 1j * data['S22 (Imag)']

# Plot the S-parameters
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(frequency, 20 * np.log10(np.abs(S11)))
plt.title('S11')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')

plt.subplot(2, 2, 2)
plt.plot(frequency, 20 * np.log10(np.abs(S21)))
plt.title('S21')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')

plt.subplot(2, 2, 3)
plt.plot(frequency, 20 * np.log10(np.abs(S12)))
plt.title('S12')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')

plt.subplot(2, 2, 4)
plt.plot(frequency, 20 * np.log10(np.abs(S22)))
plt.title('S22')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')

plt.tight_layout()
plt.show()
```
